## Step 5- Complete the fundamental data quality and consistency checks 

### 5.1 Objective: Checking for mixed data types

```python
for col in df_eco_profile.columns.tolist():
  weird = (df_eco_profile[[col]].applymap(type) != df_eco_profile[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_eco_profile[weird]) > 0:
    print (col)
```

```python
### Grouping by certain description criterion which are needed to answer the exploratory questions

grouped_df = df_eco_profile.groupby('Description')
descriptions_to_include = ['Retirement and other']

# Print unique values in 'Description' column
print("Unique Values in 'Description' column:", df_eco_profile['Description'].unique())

# Print intermediate results for debugging
for name, group in grouped_df:
    print(f"Group '{name}': {group}")

filtered_groups = grouped_df.filter(lambda group: any(category in descriptions_to_include for category in group['Description']))
print(filtered_groups)

grouped_df = df_eco_profile.groupby('Description')

descriptions_to_include = ['  Retirement and other']

filtered_groups = grouped_df.filter(lambda group: group['Description'].iloc[0] in descriptions_to_include)

print(filtered_groups)
```
